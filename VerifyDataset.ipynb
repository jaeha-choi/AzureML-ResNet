```python
from util.tiny_imagenet import TinyImagenet
from util.tiny_imagenet_val import TinyImagenetVal
from torch.utils.data import DataLoader

DATASET_LOCATION = '../tiny-imagenet-200'
IMAGE_SIZE = 64
BATCH_SIZE = 4 * 4
SHUFFLE_DATA = True;

train_dataset = TinyImagenet(DATASET_LOCATION, img_crop_size=IMAGE_SIZE)
train_dataset.save_all_dict()
train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=SHUFFLE_DATA)
val_dataset = TinyImagenetVal(DATASET_LOCATION, img_crop_size=IMAGE_SIZE)
val_dataloader = DataLoader(val_dataset, batch_size=BATCH_SIZE)
```

```python
import matplotlib.pyplot as plt
import numpy as np

count = 0
```

```python
plt.figure(figsize = (16, 6))
fig, ax = plt.subplots(4, 4)
fig.tight_layout()

(img, label) = next(iter(train_dataloader))
for x in range(4):
  for y in range(4):
    index = [4 * x + y]
    sample_image = np.transpose(np.squeeze(img[index], axis=0), axes=[1, 2, 0]) / 255
    sample_id = np.squeeze(label[index], axis=0).item()
    sample_name = train_dataset.get_class_name(sample_id)

    ax[x, y].imshow(sample_image)
    ax[x, y].text(x=10, y=-10, s=str(sample_id)+":"+sample_name, fontsize=12)
plt.show()
```

```python
plt.figure(figsize = (16, 6))
fig, ax = plt.subplots(4, 4)
fig.tight_layout()

(img, label) = next(iter(val_dataloader))
for x in range(4):
  for y in range(4):
    index = [4 * x + y]
    sample_image = np.transpose(np.squeeze(img[index], axis=0), axes=[1, 2, 0]) / 255
    sample_id = np.squeeze(label[index], axis=0).item()
    sample_name = val_dataset.get_class_name(sample_id)

    ax[x, y].imshow(sample_image)
    ax[x, y].text(x=10, y=-10, s=str(sample_id)+":"+sample_name, fontsize=12)
plt.show()
```
